<h3>필요한 라이브러리 임포트</h3>

```python
import os
import json
import pandas as pd
import math
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import time

# 폰트 설정
font_path = os.path.join(os.getcwd(), 'NanumGothic.ttf')
font_manager.fontManager.addfont(font_path)
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

<h3>필터링 및 추출 관련 함수</h3>

```python
# 전환 쌍 추출 함수
def extract_transition_pairs(df):
    transition_pairs = []
    
    # 'MOVE_TO_FOREGROUND' 타입의 데이터만 추출
    user_data = df[df['type'] == 'MOVE_TO_FOREGROUND'].reset_index(drop=True)
    
    for i in range(len(user_data) - 1):
        # 연속된 두 앱이 다를 때 전환 쌍으로 추가
        if user_data.at[i, 'name'] != user_data.at[i + 1, 'name']:
            transition_pairs.append((user_data.at[i, 'name'], user_data.at[i + 1, 'name']))
    return transition_pairs

# 사용 빈도를 추출하여, 내림차순으로 앱을 정렬하는 함수
def sort_frequent_apps(df):
    # 'MOVE_TO_FOREGROUND' 타입의 데이터에서 앱 이름별 사용 빈도 계산
    app_usage_frequency = df[df['type'] == 'MOVE_TO_FOREGROUND']['name'].value_counts().reset_index(name='frequency')
    app_usage_frequency.columns = ['app_name', 'frequency']
    
    # 사용 빈도 내림차순으로 정렬한 앱 이름 리스트 반환
    return app_usage_frequency.sort_values(by='frequency', ascending=False)['app_name'].tolist()
```

<h3>거리 계산 관련 함수</h3>

```python
# 거리 계산 함수
def calculate_distance(x, y):
    return math.sqrt((x - 2)**2 + (y - 1)**2)

# 좌표 순서 생성 함수 (한 화면에 배치될 시퀀스를 생성하는 함수)
def generate_sequence(base_coord): 
    
    # 한 화면에는 가로4개 x 세로5개의 앱이 배치되므로 (0,0)을 좌하단으로 하는 4x5의 좌표를 생성함
    coordinates = [(x, y) for x in range(4) for y in range(5)] 
    
    # base_coord로 부터 거리가 같은 좌표들 중에서는 x 좌표가 큰 값이 우선하고, 그 다음으로는 y 좌표가 작은 값을 우선하여 정렬
    coordinates = sorted(coordinates, key=lambda pos: (calculate_distance(pos[0], pos[1]), -pos[0], pos[1]))
    return coordinates
```

<h3>앱 배치 관련 함수</h3>

```python
# 앱 배치 함수
def place_apps(most_frequent_apps, rules, initial_position, page_size):
    app_lists_per_page = []  # 각 페이지별 앱 리스트를 저장할 리스트
    current_page = []  # 현재 페이지에 배치된 앱을 저장
    placed_apps = set()  # 이미 배치된 앱을 기록할 집합

    # 가장 빈번하게 사용되는 앱부터 위치에 배치
    while most_frequent_apps:
        app = most_frequent_apps.pop(0)  # 가장 자주 사용되는 앱을 하나 꺼내옴
        if app in placed_apps:  # 이미 배치된 앱이면 건너뛰기
            continue
        placed_apps.add(app)  # 앱을 배치된 앱 목록에 추가
        current_page.append(app)  # 현재 페이지에 앱 추가

        # 연관 규칙을 통해 관련 앱을 찾고 배치
        related_apps = rules[rules['antecedents'] == frozenset([app])]['consequents']
        for related in related_apps:
            for rel_app in related:
                if rel_app not in placed_apps and len(current_page) < page_size:
                    placed_apps.add(rel_app)  # 관련 앱을 배치된 앱 목록에 추가
                    current_page.append(rel_app)  # 현재 페이지에 앱 추가
                    if rel_app in most_frequent_apps:
                        most_frequent_apps.remove(rel_app)  # 앱을 배치 했으므로, most_frequent_apps에도 해당 앱이 있었다면 제거

        # 현재 페이지가 가득 찼거나 더 이상 배치할 앱이 없으면 새 페이지 시작
        if len(current_page) == page_size or not most_frequent_apps:
            app_lists_per_page.append(current_page)
            current_page = []

    return app_lists_per_page

# 앱 배치 시각화 함수
def visualize_app_coordinates(user_id, apps, coordinate_sequence, page_size):

    # 페이지 넘버 초기화
    page_number = 1
    
    # 앱 목록을 넘겨받은 좌표 순서대로 좌표에 매핑
    apps_coordinates = {app: coordinate_sequence[i % page_size] for i, app in enumerate(apps)}
    
    # 각 페이지마다 시각화
    for idx, (app, coord) in enumerate(apps_coordinates.items()):
        
        # 좌표별 앱 시각화
        x, y = coord
        plt.scatter(x, y, label=app, s=100)
        
        # 페이지 내 마지막 앱 또는 전체 목록의 마지막 앱 처리
        if (idx + 1) % page_size == 0 or (idx + 1) == len(apps):
            plt.title(f'User{user_id}의 페이지{page_number}')
            plt.xlim(-0.5, 3.5)
            plt.ylim(-0.5, 4.5)
            plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
            plt.tight_layout()
            plt.show()
            page_number +=1
```

<h3>평가 관련 함수</h3>

```python
# 화면에 띄워진 앱 로그만 선별하는 함수
def filter_foreground_apps(data):
    
    # type : MOVE_TO_FOREGROUND인 경우만 필터링 (즉, 화면에 앱이 띄워진 경우만 선별)
    app_sequence = [entry['name'] for entry in data if entry['type'] == 'MOVE_TO_FOREGROUND']
    return app_sequence

# 평가시 비용 계산을 위한 함수
def get_cost(app_list, user_list, dist):
    total_cost = 0 # 해당 사용자의 총 비용
    page_now = 0 # 시작 페이지
    
    for target_app in user_list: # 사용한 앱마다
        for page_index, page in enumerate(app_list): # 각 페이지를 순회하며
            if target_app in page: # 해당 앱을 페이지에서 찾음
                
                # 페이지내 해당 앱의 인덱스
                app_index = page.index(target_app)
                
                # 거리와 페이지 인덱스를 고려하여 해당 앱 사용시 드는 비용 계산
                if page_index != page_now :
                    app_cost = dist[app_index] + (abs(page_index - page_now) * 3) # 페이지 변경시 1페이지당 3씩 비용을 더함
                    page_now = page_index
                else:
                    app_cost = dist[app_index]
                    
                total_cost += app_cost
                break
                
    return total_cost
```

<h3>메인 함수</h3>

```python
if __name__ == "__main__":
    
    # 실행 시간 측정
    start_time = time.time()

    # 배치 완료 후, 각 앱 리스트들을 파일로 저장할 폴더
    result_folder = "../apriori_results" 
    if not os.path.exists(result_folder):
        os.makedirs(result_folder)

    page_size = 20 # 한 페이지의 크기 (앱 개수)
    initial_position = (2, 1)  # 손가락이 가장 편히 오는 점 (거리 측정 시 기준점)

    # user1 _train.json부터 user33_train.json까지 처리
    for user_id in range(1, 34):

        # 데이터 로드
        folder_path = "../user_data_train"
        file_name = f"user{user_id}_train.json"
        file_path = os.path.join(folder_path, file_name)
        if not os.path.exists(file_path):
            print(f"파일을 찾을 수 없습니다: {file_path}")
            continue

        with open(file_path, 'r', encoding='utf-8') as f:
            data = json.load(f)

        df = pd.DataFrame(data)

        # 전환 쌍 추출
        transition_pairs = extract_transition_pairs(df)

        # 전환 쌍 데이터프레임 생성
        transition_df = pd.DataFrame(transition_pairs, columns=['app_from', 'app_to'])

        # 피벗 테이블 형식으로 변환
        pivot_df = transition_df.pivot_table(index='app_from', columns='app_to', aggfunc='size', fill_value=0)

        # 장바구니 형식의 데이터프레임 생성, Apriori 알고리즘 적용
        basket = pivot_df.apply(lambda x: x.map(lambda y: y > 0)).astype(bool)

        # 모든 앱 전환쌍 중에서 10% 이상을 차지하는 앱 조합은 빈발 아이템셋으로 선정 
        frequent_itemsets = apriori(basket, min_support=0.1, use_colnames=True)

        # 두 앱이 독립적으로 사용될 때보다 함께 사용될 가능성이 더 높은 경우(min_threshold=1)만 선택 
        rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

        # 사용 빈도가 높은 앱부터 정렬하여 most_frequent_apps에 저장
        most_frequent_apps = sort_frequent_apps(df)

        # 앱 배치 순서 생성
        positions = generate_sequence(initial_position)

        # 앱 배치
        result_positions = place_apps(most_frequent_apps, rules, initial_position, page_size)

        # 배치된 앱 출력
        print(f'user{user_id}의 최종 앱리스트 : ')
        for page in result_positions:
            print(page)  # 각 페이지별 앱 리스트 출력

        # 배치된 앱 리스트를 파일로 저장
        result_file_path = os.path.join(result_folder, f"user{user_id}.json")
        with open(result_file_path, "w", encoding="utf-8") as result_file:
            json.dump(result_positions, result_file, ensure_ascii=False, indent=4)

        print(f"\n### User{user_id}의 최적화된 앱 리스트가 '{result_file_path}'에 저장되었습니다. ###\n\n")
        
    # 실행 시간 출력
    end_time = time.time()
    print("최적화 실행 시간:", end_time - start_time, "초")
```

<h3>사용자별 앱 배치 결과를 시각화</h3>

```python
# 최적화된 결과 파일 목록 불러오기
result_folder = '../apriori_results'

# 좌표 리스트
coordinate_sequence = generate_sequence((2,1))

# 파일 목록을 불러와서 user_id를 기준으로 정렬
result_files = sorted(
    [f for f in os.listdir(result_folder) if f.endswith('.json')],
    key=lambda x: int(x.split('user')[1].split('.json')[0])
)

for result_file in result_files:
    user_id = int(result_file.split('user')[1].split('.json')[0])  # 파일명에서 user_id 추출
    file_path = os.path.join(result_folder, result_file)  # JSON 파일 로드
    if not os.path.exists(file_path):
        print(f"파일을 찾을 수 없습니다: {file_path}")
        continue
    
    with open(file_path, 'r', encoding='utf-8') as file:
        data = json.load(file)
        
    # 앱 목록을 1차원 배열로 변환
    apps = [app for sublist in data for app in sublist]
    
    # 사용자별 앱 좌표 시각화
    visualize_app_coordinates(user_id, apps, coordinate_sequence, page_size)
```
