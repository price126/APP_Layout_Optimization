<h3>필요한 라이브러리 import</h3>

```python
import json
import os
from collections import defaultdict
import math
```

<h3>사용자별로 user1 ~ user33 까지 분할</h3>

```python
# user_data 폴더 생성
if not os.path.exists('user_data'):
    os.makedirs('user_data')

# 데이터 로드
with open('../스마트폰_로그_데이터셋.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# 사용자별 데이터 분류
user_data = defaultdict(list)
for entry in data:
    user_data[entry['subject_id']].append(entry)

# 사용자별 데이터 저장
user_number = 1
for user_id, entries in user_data.items():
    filename = f'user_data/user{user_number}.json'
    with open(filename, 'w', encoding='utf-8') as file:
        json.dump(entries, file, ensure_ascii=False, indent=4)
    print(f'{filename} 파일이 저장되었습니다.')
    user_number += 1
```

<h3>각 user.json 별 소유한 데이터 개수</h3>

```python
# 'user_data' 폴더 내의 모든 파일을 읽기
file_list = os.listdir('user_data')

# 파일 이름에 따라 숫자 부분을 기준으로 정렬
# 예: "user1.json"에서 숫자를 추출하여 정수로 변환, 정렬 기준으로 사용
file_list_sorted = sorted(file_list, key=lambda x: int(x.replace('user', '').replace('.json', '')))

# 정렬된 파일 리스트를 순회하며 각 파일의 데이터 개수 세기
for filename in file_list_sorted:
    # 파일 경로 조합
    file_path = os.path.join('user_data', filename)
    
    # JSON 파일 열기 및 데이터 로드
    with open(file_path, 'r', encoding='utf-8') as file:
        data = json.load(file)
    
    # 데이터 개수 출력
    print(f'{filename}: {len(data)}개')
```

<h3>데이터 개수가 각 유저마다 충분한 것 같으니 5:5으로 분할</h3>

```python
# 폴더 생성 함수
def create_folder_if_not_exists(folder_name):
    if not os.path.exists(folder_name):
        os.makedirs(folder_name)

# JSON 파일 분할 함수
def split_json_file(file_path, train_folder, test_folder):
    with open(file_path, 'r', encoding='utf-8') as file:
        data = json.load(file)

    total_entries = len(data)
    split_index = math.ceil(total_entries * 0.5)

    train_data = data[:split_index]
    test_data = data[split_index:]

    base_name = os.path.basename(file_path)
    file_name, file_ext = os.path.splitext(base_name)

    train_file_path = os.path.join(train_folder, f"{file_name}_train{file_ext}")
    test_file_path = os.path.join(test_folder, f"{file_name}_test{file_ext}")

    with open(train_file_path, 'w', encoding='utf-8') as train_file:
        json.dump(train_data, train_file, ensure_ascii=False, indent=4)

    with open(test_file_path, 'w', encoding='utf-8') as test_file:
        json.dump(test_data, test_file, ensure_ascii=False, indent=4)

    print(f"{base_name} 파일을 {len(train_data)}개의 훈련 데이터와 {len(test_data)}개의 테스트 데이터로 분할했습니다.")

input_folder = 'user_data'
train_folder = 'user_data_train'
test_folder = 'user_data_test'

create_folder_if_not_exists(train_folder)
create_folder_if_not_exists(test_folder)

for file_name in sorted(os.listdir(input_folder)):
    if file_name.endswith('.json'):
        file_path = os.path.join(input_folder, file_name)
        split_json_file(file_path, train_folder, test_folder)
```

<h3><b>subject_id를 1 ~ 33으로 통일시킴</b></h3>

```python
# 폴더 내의 모든 파일을 순회하며 subject_id를 수정하는 함수
def convert_subject_id(folder_path):
    # 파일 종류 확인 (train 또는 test)
    file_type = 'train' if 'train' in folder_path else 'test'
    
    for filename in os.listdir(folder_path):
        if filename.endswith('.json'):
            # 파일명에서 user 번호 추출 (예: user1_train.json 에서 1 추출)
            user_number = filename.split('_')[0].replace('user', '')

            # JSON 파일 열기
            file_path = os.path.join(folder_path, filename)
            with open(file_path, 'r', encoding='utf-8') as file:
                data = json.load(file)

            # 각 엔트리의 subject_id 수정
            for entry in data:
                entry['subject_id'] = user_number

            # 수정된 데이터로 파일 다시 쓰기
            with open(file_path, 'w', encoding='utf-8') as file:
                json.dump(data, file, ensure_ascii=False, indent=4)

    print(f"모든 {file_type} 파일의 subject_id가 수정되었습니다.")

    
train_folder = 'user_data_train'
test_folder = 'user_data_test'

convert_subject_id(train_folder)
convert_subject_id(test_folder)
```

<h3><b>시스템앱은 배치에 쓰이는 일반적인 앱이 아니므로 모두 제거</b></h3>

```python
# 폴더 내의 모든 파일을 순회하며 is_system_app = 1 인 엔트리를 제거하는 함수
def remove_system_apps(folder_path):
    # 파일 이름의 접미사 결정 (_train 또는 _test)
    suffix = 'train' if 'train' in folder_path else 'test'
    
    for i in range(1, 34):  # user1 ~ user33까지
        file_path = os.path.join(folder_path, f'user{i}_{suffix}.json')

        try:
            # 파일을 열고 JSON 데이터를 로드
            with open(file_path, 'r', encoding='utf-8') as file:
                data = json.load(file)

            # "is_system_app"가 1인 엔트리를 제거
            filtered_data = [entry for entry in data if entry.get("is_system_app") != 1]

            # 변경된 데이터를 같은 파일에 덮어쓰기
            with open(file_path, 'w', encoding='utf-8') as file:
                json.dump(filtered_data, file, ensure_ascii=False, indent=4)

            print(f'{file_path} 처리 완료')

        except Exception as e:
            print(f'{file_path} 처리 중 오류 발생: {e}')

    print(f"{folder_path} 내의 파일들에서 시스템 앱 제거 완료")

train_folder = 'user_data_train'
test_folder = 'user_data_test'

remove_system_apps(train_folder)
remove_system_apps(test_folder)
```

<h3><b>앱 이름이 비어있는 로그 제거</b></h3>

```python
# 폴더 내의 모든 파일을 순회하며 name = "" 인 엔트리를 제거하는 함수
def remove_empty_names(folder_path):
    # 파일 이름의 접미사 결정 (_train 또는 _test)
    suffix = 'train' if 'train' in folder_path else 'test'
    
    for i in range(1, 34):  # user1 ~ user33까지
        file_path = os.path.join(folder_path, f'user{i}_{suffix}.json')

        try:
            # 파일을 열고 JSON 데이터를 로드
            with open(file_path, 'r', encoding='utf-8') as file:
                data = json.load(file)

            # name이 ""인 엔트리를 제거
            filtered_data = [entry for entry in data if entry.get("name") != ""]

            # 변경된 데이터를 같은 파일에 덮어쓰기
            with open(file_path, 'w', encoding='utf-8') as file:
                json.dump(filtered_data, file, ensure_ascii=False, indent=4)

            print(f'{file_path} 처리 완료')

        except Exception as e:
            print(f'{file_path} 처리 중 오류 발생: {e}')

    print(f"{folder_path} 파일들에서 name이 비어있는 로그 제거 완료")


train_folder = 'user_data_train'
test_folder = 'user_data_test'

remove_empty_names(train_folder)
remove_empty_names(test_folder)
```
